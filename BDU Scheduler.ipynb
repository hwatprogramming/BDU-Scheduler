# CSV Extraction

```python
def extract_table(csv_data):
    rows = csv_data.shape[0]
    cols = csv_data.shape[1]
    table = np.zeros((rows,cols-1), dtype=int)
    for i in range(rows):
        for j in range(1,cols,1):
            print(j,i)
            table[i][j-1] = csv_data[j][i]
    return table
```

```python
def extract_names(csv_data):
    rows = csv_data.shape[0]
    
    longest_length = len("Morning Check")
    for j in range(rows):
        if len(csv_data[0][j]) > longest_length:
            longest_length = len(csv_data[0][j])
    string_type = "S"+str(longest_length)
    #print(string_type)
    
    names = np.empty(rows, dtype=string_type)
    for i in range(rows):
        #print (csv_data[0][i])
        names[i] = csv_data[0][i]
        
    names_str = names.astype(str)
    return names_str
```

```python
def check(timetable, names, goals):
    
    num_people = len(timetable)
    num_days = len(timetable[0])
    
    morning_req = 2
    night_req = 3
    
    morning_check = np.zeros(num_days, dtype=int)
    night_check = np.zeros(num_days, dtype=int)
    
    print("Number of people in this schedule:",num_people)
    print("Number of days in this schedule:",num_days)
    print()
    
    for i in range(num_days):
        morning_count = 0
        night_count = 0
        for j in range (num_people):
            if timetable[j][i] == 1:
                morning_count += 1    
            if timetable[j][i] == 2:
                night_count += 1
        if morning_count == morning_req:
            morning_check[i] = 1
        if night_count == night_req:
            night_check[i] = 1
            
    score = print_stylized_table(timetable, names, morning_check, night_check, goals)
    return score
```

```python
def longest_length(names):
    longest_length = 1
    for j in range(len(names)):
        if len(str(names[j])) > longest_length:
            longest_length = len(str(names[j]))
    return longest_length
```

```python
def difference_array(scores, goals):
    num_people = len(scores)
    difference = np.zeros(num_people, dtype=int)
    for i in range(num_people):
        difference[i] = goals[i] - scores[i]
    return difference
```
